```python
import re
import collections
```

```python
path = 'data/hsa00001.metab.keg.txt'
```

```python
! head $path
```

```python
p = ''
p2g = collections.defaultdict(set)
with open(path, 'r') as f:
    for l in f: 
        s = l[1:].strip()
        if l[0] == 'C':
            s = re.sub(r'\[PATH:.+\]', '', s)
            s = re.sub('\d{5}', '', s).strip()
            p = s
#             print(s)
        if l[0] == 'D':
            g = re.search(' (.+);', s)
            if g:
                p2g[p].add(g[1])
            else:
                print(s)
```

```python
print(p2g)
```

```python
with open("data/hsa00001.metab.keg.gmt", "w") as f:
    for p in p2g:
        g = p2g[p]
        f.write("{}\n".format("\t".join([p]+list(g))))
```
